```python
import pandas as pd
```

```python
dataset=pd.read_csv("https://raw.githubusercontent.com/YBI-Foundation/Dataset/main/Diabetes.csv")
dataset.head(10)
```

```python
dataset.columns
```

```python
dataset.info()
dataset.describe()
```

```python
features=["pregnancies","glucose","diastolic","triceps","insulin","bmi","dpf","age"]
x=dataset[features]
y=dataset['diabetes']
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,train_size=0.65,random_state=2529)
x_train.shape, x_test.shape, y_train.shape, y_test.shape
```

```python
from sklearn.linear_model import LogisticRegression
model1=LogisticRegression(max_iter=2000)
```

```python
model1.fit(x_train,y_train)
prediction1=model1.predict(x_test)
prediction1
```

```python
from sklearn.neighbors import KNeighborsClassifier
model2=KNeighborsClassifier()
model2.fit(x_train,y_train)
prediction2=model2.predict(x_test)
prediction2
```

```python
from sklearn.tree import DecisionTreeClassifier
model3=DecisionTreeClassifier()
model3.fit(x_train,y_train)
prediction3=model3.predict(x_test)
prediction3
```

```python
from sklearn.metrics import accuracy_score
print("For Logistic Regression:",accuracy_score(y_test,prediction1))
print("For KNN Classification:",accuracy_score(y_test,prediction2))
print("For Decision Tree Classifier:",accuracy_score(y_test,prediction3))
```
